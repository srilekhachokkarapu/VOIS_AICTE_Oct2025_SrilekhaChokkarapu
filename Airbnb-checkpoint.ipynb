```python
# Import essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style for better aesthetics
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6) # Default figure size

# --- Data Loading ---
# NOTE: The file is a CSV, so we use pd.read_csv.
# If you upload an actual Excel file (.xlsx), you would need to use:
# df = pd.read_excel('Your_File_Name.xlsx')

file_name = '1730285881-Airbnb_Open_Data.xlsx - in.csv'

try:
    # Load the dataset using the exact file name
    df = pd.read_excel('1730285881-Airbnb_Open_Data.xlsx')
    print(f"✅ Successfully loaded {file_name}. Initial shape: {df.shape}")
except FileNotFoundError:
    print(f"❌ Error: The file '{file_name}' was not found.")
    print("Please ensure the file is uploaded to the same directory as your Jupyter Notebook or Colab session.")
    # Exit or stop execution if the file isn't loaded
    raise
```

```python

# Display initial information
print("\n--- Initial Dataset Info ---")
df.info()
```

```python

print("\n--- First 5 Rows ---")
print(df.head())
```

```python
# --- 2.1 Standardize Column Names ---
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('.', '', regex=False)
print("\n--- Column Names Cleaned ---")
print(df.columns.tolist())

# --- 2.2 Clean and Convert Numerical/Currency Columns ---

def clean_and_convert_currency(series):
    """Removes currency symbols and commas, then converts the series to float."""
    # Convert to string first to handle potential mixed types (e.g., NaN)
    return series.astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)

# Apply cleaning to 'price' and 'service_fee'
df['price'] = clean_and_convert_currency(df['price'])
df['service_fee'] = clean_and_convert_currency(df['service_fee'])

# --- 2.3 Handle Missing Values (NaN) ---

# Check for null values
print("\n--- Missing Value Count (Before Cleaning) ---")
print(df.isnull().sum().sort_values(ascending=False).head(10))

# Drop columns with high sparsity or low analytical value for a quick EDA
df.drop(columns=['house_rules', 'license', 'country', 'country_code'], inplace=True, errors='ignore')

# Impute 'reviews_per_month': Assume 0 for listings with no reviews/last review date
df['reviews_per_month'].fillna(0, inplace=True)

# Convert 'last_review' to datetime and fill NaT with a sentinel date (if needed), 
# but for EDA we'll just drop rows if other critical data is missing.
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Drop rows where critical identifying info or location data is missing
df.dropna(subset=['host_name', 'name', 'neighbourhood_group', 'neighbourhood', 'lat', 'long'], inplace=True)

# --- 2.4 Handle Outliers ---
# Outliers in price can skew visualization. Let's cap the price for visualization.
# We'll calculate the IQR and define a reasonable upper bound.
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
price_cap = Q3 + 3 * IQR # A more generous cap for analysis

# Filter the DataFrame
df_cleaned = df[df['price'] > 0] # Remove zero-priced listings
df_cleaned = df_cleaned[df_cleaned['price'] <= price_cap].copy()

print(f"\n✅ Data Cleaning Complete.")
print(f"Rows remaining after cleaning (Price > 0 and <= ${price_cap:.0f}): {df_cleaned.shape[0]} rows.")
print("\n--- Final Cleaned Data Info ---")
df_cleaned.info()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['price'], bins=50, kde=True, color='#00a699')
plt.title('Distribution of Listing Prices (USD)', fontsize=16, fontweight='bold')
plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Number of Listings', fontsize=12)
plt.show()

print(f"\n--- Key Price Statistics ---")
print(df_cleaned['price'].describe().to_string())
```

```python
# 1. Listings by Neighborhood Group (Count Plot)
plt.figure(figsize=(12, 6))
sns.countplot(y='neighbourhood_group', data=df_cleaned, 
              order=df_cleaned['neighbourhood_group'].value_counts().index, 
              palette='viridis')
plt.title('Listing Popularity by Neighborhood Group', fontsize=16, fontweight='bold')
plt.xlabel('Number of Listings', fontsize=12)
plt.ylabel('Neighborhood Group', fontsize=12)
plt.show()

# 2. Average Price by Neighborhood Group (Bar Plot)
avg_price_geo = df_cleaned.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='neighbourhood_group', y='price', data=avg_price_geo, palette='coolwarm')
plt.title('Average Price by Neighborhood Group', fontsize=16, fontweight='bold')
plt.xlabel('Neighborhood Group', fontsize=12)
plt.ylabel('Average Price ($)', fontsize=12)
plt.show()
```

```python
# Room Type Count
plt.figure(figsize=(8, 5))
sns.countplot(x='room_type', data=df_cleaned, palette='Set1')
plt.title('Distribution of Listings by Room Type', fontsize=16, fontweight='bold')
plt.xlabel('Room Type', fontsize=12)
plt.ylabel('Number of Listings', fontsize=12)
plt.show()

# Price Distribution by Room Type (Box Plot)
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df_cleaned, palette='Set1')
plt.title('Price Distribution by Room Type', fontsize=16, fontweight='bold')
plt.xlabel('Room Type', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.show()
```

```python
# Calculate correlation matrix for key numerical variables
core_features = ['price', 'service_fee', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365']
corr_matrix = df_cleaned[core_features].corr()

plt.figure(figsize=(8, 7))
sns.heatmap(corr_matrix, annot=True, cmap='cividis', fmt=".2f", linewidths=.5, cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Correlation Matrix of Key Numerical Features', fontsize=16, fontweight='bold')
plt.show()

# Insight: Price vs. Number of Reviews (to visualize the correlation)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='number_of_reviews', data=df_cleaned, alpha=0.5, hue='room_type', palette='deep')
plt.title('Price vs. Number of Reviews', fontsize=16, fontweight='bold')
plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.legend(title='Room Type')
plt.show()
```
